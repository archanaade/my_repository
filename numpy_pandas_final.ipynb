NUMPY

```python
import numpy as np #np is the alias name
```

```python
my_list=[3,4,6,7,9]
```

```python
type(my_list)
```

```python
#converting list into array
arr=np.array(my_list)
```

```python
print(arr)
```

```python
type(arr) #nd stands for ndimentional and here value of 'n' is 1, 1D array
```

```python
#multinested array/2D arrays
my_list_1=[3,4,6,7,9]
my_list_2=[6,7,0,1,9]
my_list_3=[2,5,8,3,4]

arr_new=np.array([my_list_1,my_list_2,my_list_3])
```

```python
print(arr_new)
```

```python
type(arr_new) #here value of n is 2, 2D array
```

```python
arr_new.shape #here 3 indicates number of rows and 5 indicates number of columns
```

```python
arr_new.reshape(5,3)
```

```python
arr_new.reshape(4,5)
```

```python
#generating numbers using numpy
np.arange(1,20,1) #here we have a range of 1-20(20 being exclusive),
#and step size of 1 which means uniform difference of '1' throughout
```

```python
np.arange(1,20,5) #step size is 5
```

```python
np.linspace(1,20,50) #here 50 indicates the count of numbers between 1-20 range
```

```python
np.linspace(1,20,80) #here 80 indicates the count of numbers between 1-20 range
```

```python
#random numbers
np.random.rand(3,5) #here 3 indicates the number of rows and 5 indicates the number of columns
#here by default the range is between 0-1(1 being exclusive)
#everytime we run the code cell, we will get randomly generated numbers
```

```python
np.random.randint(1,20,4) #here 1-20 is the range(20 being exclusive) and 4 indicates the count of random numbers needed in the range
#everytime we run the code cell, we will get randomly generated numbers
```

```python
np.random.randint(10,20,3)
```
